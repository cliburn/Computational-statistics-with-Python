### Conjugate gradient example for CUDA
<http://www.tsc.uc3m.es/~miguel/MLG/adjuntos/slidesCUDA.pdf>

```python
def conjGrad (A, b, tol=1.0e-3):
    N = len(b)
    x = np.zeros_like(b)
    r = b - np.dot(A, x)
    p = r.copy()
    for i in range(N):
        z = np.dot(A ,p)
        alpha = np.dot(p, r) / np.dot(p, z)
        x = x + alpha*p
        r = b - np.dot(A, x)
        if (np.sqrt(np.dot(r, r))) < tol:
            break
        else :
            beta = -np.dot(r, z) / np.dot(p , z)
        p = r + beta*p
    print "Iterations required on CPU:" , i
    return x
```

```python
import time as t

# generate GP
x = np.arange(-5 ,5 ,0.01).reshape(-1 ,1)
N = len(x)
K = np.exp(-0.5/0.7*(np.dot(x*x ,np.ones((1, N))) + 
                     np.dot(np.ones((N, 1)), (x*x ).T)-2*np.dot (x, x.T)))
Kn = K + np.eye(N)
L = np.linalg.cholesky(Kn)
y = np.dot(L ,np.random.randn(N))

# K = K.astype(np.float32)
# Kn = Kn.astype(np.float32) 
# y = y.astype(np.float32) ;

plt.figure(figsize=(10,5))
s = t.time()
alpha = np.linalg.solve(Kn , y)
mu = np.dot(K, alpha)
print "CPU: Numpy found solution in %f secs" % ( t.time()-s )
plt.subplot(1,2,1)
plt.plot (x , y ,'bx' ,x ,mu,'r');

s = t.time()
alpha = conjGrad(Kn, y)
mu = np.dot(K, alpha)
print "CPU: Conjugate gradient found solution in %f secs" % ( t.time()-s )
plt.subplot(1,2,2)
plt.plot (x , y ,'bx' ,x ,mu,'r');
```
