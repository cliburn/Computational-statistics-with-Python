````python
x = fcm.loadFCS('data/2311.fcs', auto_comp=True, transform='logicle')
````

````python
zip(range(100), x.channels)
````

Description of markers
----

- FSC - Forward scatter (function of cell size)
- SSC - Side scatter (function of cell granularity)
- KI67 - High in activated/dividing cells
- CD8 - Marker for cytotoxic T cells
- HELOPS - Marker for "naive" Tregs
- CD4 - Marker for helper T cells (superset of Tregs)
- CD45RA - Marker for memory T cells
- CTLA4 - Inhibitory co-stimulatory moleucle (high in Tregs)
- CD3 - Marker for T cells (superset of CD4 and CD8 T cells)
- YARD - Live/dead markers - cells high in YARD are likely to be dead cells
- CD127 - Marker for Tregs (low in Tregs)
- FOXP3 - Marker for Tregs (high in Tregs)
- CD25 - Marker for Tregs (high in Tregs)
- Time - Time of event acquisition - ussd for Quality control

````python
channels = x.channels
````

````python
dims = [(0,3), (1,2), (12,13), (9,7), (9, 16), (16,15), (10,11), (8,6), (14,15)]

ncols = 3
nrows = 3
plt.figure(figsize=(nrows*3, ncols*3))

npts = 10000

for k, dim in enumerate(dims):
    plt.subplot(nrows, ncols, k+1)

    z = bilinear_interpolate(x[:npts, dim[0]], x[:npts, dim[1]])
    plt.scatter(x[:npts, dim[0]], x[:npts, dim[1]], s=1, c=z, edgecolors='none')

    plt.xlabel(channels[dim[0]], fontsize=16)
    plt.ylabel(channels[dim[1]], fontsize=16)
    if 'FSC' in channels[dim[0]]  or 'SSC' in channels[dim[0]]:
        plt.xlim([0, 265000])
    else:
        plt.xlim([0, 100000])
    if 'FSC' in channels[dim[1]]  or 'SSC' in channels[dim[1]]:
        plt.ylim([0, 265000])
    else:
        plt.ylim([0,100000])
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
````
