```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
%precision 4
import os, sys, glob
```

```python
def f(x):
    return 2.5*x*x - 30*x + 100

x = np.linspace(0, 10, 100)
y = f(x)

a = 1
b = 9
ef = (f(a) + f(b))/2.0
fe = f((a+b)/2.0)

plt.xkcd()
plt.plot(x, y)
plt.ylim([0, 90])
plt.axvline(a, 0, f(a)/90.0, c='green', ls='dashed')
plt.axvline(b, 0, f(b)/90.0, c='green', ls='dashed')
plt.xticks([a, b, (a+b)/2.0], ['a', 'b', 'E[x]'], fontsize=20)
plt.axhline(f(a), 0, a/10.0, c='green', ls='dashed')
plt.axhline(f(b), 0, b/10.0, c='green', ls='dashed')
plt.yticks([f(a), f(b), f((a+b)/2.0), ef], ['f(a)', 'f(b)', 'f(E[x])', 'E[f(x)]'], fontsize=20)
plt.axvline((a+b)/2.0, 0, fe/90.0, c='red')
plt.axhline(f((a+b)/2.0), 0, (a+b)/20.0, c='red')
plt.title("Jensen's inequality for convex function", fontsize=20);
plt.text((a+b)/2.0-1, f((a+b)/2.0)+40, 'E[f(x)] $\geq$ f(E[x])', fontsize=20)
plt.savefig('figs/jensen.png')
plt.tight_layout()
```
