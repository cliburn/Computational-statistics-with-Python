```python
from numba import jit
import bitey
import numpy as np
%load_ext cythonmagic
%load_ext fortranmagic
```

```python
%%file loop.c

long f_c(long N) {
    long s = 0;
    for (long i=0; i<N; i++) {
        s += i*i;
    }
    return s;
}
```

```python
!clang -O3 -emit-llvm -c loop.c -o loop.o
```

```python
%%fortran

subroutine f_fort(s, N)
    integer*8, intent(in) :: N
    integer*8, intent(out) :: s
    integer*8 i
    s = 0
    do i = 1, N
        s = s + i*i
    end do
end 
```

```python
def f_py(N):
    s = 0
    for i in xrange(N):
        s += i*i
    return s
```

```python
f_jit = jit(f_py)
```

```python
%%cython -a
def f_cy(long N):
    cdef long s = 0
    cdef long i
    for i in xrange(N):
        s += i*i
    return s
```

```python
def f_np(N):
    xs = np.arange(N, dtype='i8')
    return np.sum(xs*xs)
```
