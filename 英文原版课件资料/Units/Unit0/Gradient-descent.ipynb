We will first do gradient descent for a linear regression function. The residulas are given by

\\[
\sum_{i=1}^{N} (y_i - (\beta_0 + \beta_1x))^2
\\]

and we have two parameters $\beta_0$ and $\beta_1$ that we need to optimize. In matrix notation, we have the residual function $f$ given by

\begin{align}
& (y - X\beta)^2 \\ 
&= (y - X\beta)^T(y - X\beta) \\
&= y^Ty -2\beta^TX^Ty + \beta^TX^TX\beta
\end{align}

Differentiating with respect to $\beta$, we get the derivative as

\begin{align}
-2X^Ty + 2(X^TX)\beta
\end{align}

We can of course equate this to zero and recovert the normal equaions

\begin{align}
\beta = (X^TX)^{-1}X^Ty 
\end{align}

but for illustrating gradient descent optimizaiotn, will just need the derivative.

```python
import os
import sys
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
%precision 4
```

Main idea
----

Given the observed data $X$ and $y$, the residual is a function only of the vector $\beta$. In this simple example, $\beta$ has 2 components, and so $r(\beta)$ is a surface that depends on $\beta_0$ and $\beta_1$ changes. We sketch the residual function surface here, and note that optimization is just trying to find the value of $\beta$ corresponding to the lowest point on this surface (where the residuals would be the smallest). Starting from any point, gradient descent simply takes a step in the opposite direction of the gradient at that point, hence going "downhill". At the minimum, the gradient is 0 and we cannot descend any furhter. 

```python
def resid(y, x, b0, b1):
    b = np.array([b0, b1])
    x = np.vstack([np.ones_like(x), x]).T
    return np.sum((y - np.dot(x, b))**2)
```

```python
b0, b1 = [5, 5]
xs = np.linspace(0, 1, 101)
ys = b0 + b1*xs + np.random.randn(len(xs))
```

```python
intervals = 10
us = np.linspace(0, 10, intervals)
vs = np.linspace(0, 10, intervals)
U, V = np.meshgrid(us, vs)
Z = np.array([resid(ys, xs, u, v) for u, v in zip(U.ravel(), V.ravel())])
Z = Z.reshape((intervals, intervals))
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(U, V, Z, )
plt.xlabel('b0', fontsize=16)
plt.ylabel('b1', fontsize=16)
plt.xlim([0,10])
plt.ylim([0,10])
plt.tight_layout()
```

The gradient descent algorithm
----

```python
def grad(xs, ys, b0, b1):
    b = np.array([b0, b1])
    x = np.vstack([np.ones_like(xs), xs]).T
    return -np.dot(x.T, ys) + np.dot(np.dot(x.T, x), b)
```

```python
def gradient_descent(xs, ys, b0_, b1_, f, fprime, learn_rate=0.001, tol=1e-6, max_iter=1000):
    b = np.array([b0_, b1_], 'float')
    delta = np.zeros_like(b)

    bs = [[b[0], b[1], f(ys, xs, b[0], b[1])]]
    for i in range(max_iter):
        delta = learn_rate*fprime(xs, ys, b[0], b[1])
        b -= delta
        bs.append([b[0], b[1], f(ys, xs, b[0], b[1])])
        if np.sqrt(np.sum(delta*delta)) < tol:
            break
    return np.array(bs)
```

```python
bs = gradient_descent(xs, ys, 10, 10, resid, grad, learn_rate=0.001)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,8))

ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(U, V, Z, )
ax.scatter(bs[:, 0], bs[:, 1], bs[:,2], c='red')
plt.xlabel('b0', fontsize=16)
plt.ylabel('b1', fontsize=16)
plt.xlim([0,10])
plt.ylim([0,10])
plt.tight_layout()
```

```python
plt.scatter(xs, ys)
for i, b in enumerate(bs[[0,1,2,4,8,16,32]]):
    plt.plot(xs, b[0] + b[1]*xs, c='red', linewidth=0.2);
plt.plot(xs, bs[-1,0] + bs[-1,1]*xs, c='red', linewidth=2)
```
