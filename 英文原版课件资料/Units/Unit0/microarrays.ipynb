### Example of quantile normalization

```python
xs = np.array([[5,4,3], [2,1,4], [3,4,6],[4,2,8]])
```

```python
xs
```

```python
from scipy.stats import rankdata
from functools import partial

ranking = partial(rankdata, method='ordinal')
```

```python
def quantile_normalize(xs):
    ranks = np.apply_along_axis(ranking, 0, xs).astype('int') - 1
    ref = np.sort(xs, 0).mean(1)
    result = ref[ranks]
    return ranks, ref, result
```

```python
ranks, ref, result = quantile_normalize(xs)
print "orignal array"
print xs
print
print "ranks"
print ranks
print
print "ref values"
print ref
print
print "normalized array"
print result
```

### Example of median polish to summarize gene expression on chip over mutltiple probes for same gene

```python
xs = np.array([61.9,54.3,56.1,50.6,
68.0,71.8,64.8,66.3,
68.8,76.6,73.1,73.9,
75.5,83.3,78.1,78.1]).reshape(4,4).T
```

```python
xs
```

```python
def median_polish(xs, niters=10):
    n, p = xs.shape
    rr = np.zeros((n,1))
    cc = np.zeros((1,p))
    resid = xs.copy()
    for i in range(niters):
        r = np.median(resid, 1).reshape(4,1)
        resid -= r
        c = np.median(resid, 0).reshape(1,4)
        resid -= c
        rr += r
        cc += c
    return resid, rr, cc, np.median(rr) + np.median(cc)
```

```python
resid, rr, cc, mu = median_polish(xs)

print "residulas\n", resid
print "row effects\n", rr.T
print "column effects\n", cc
print "average effect\n", mu
```

### PCA can be viewed as an eigendecomposition of the covariance matrix

Remember the formula for covariance

$$
\text{Cov}(X, Y) = \frac{\sum_{i=1}^n(X_i - \bar{X})(Y_i - \bar{Y})}{n-1}
$$

```python
mu = [0,0]
sigma = [[0.6,0.2],[0.2,0.2]]
n = 1000
x = np.random.multivariate_normal(mu, sigma, n).T
```

```python
cx = np.cov(x)
```

```python
m = np.array([[1,2,3],[6,5,4]])
ms = m - m.mean(1).reshape(2,1)
np.dot(ms, ms.T)/2
```

```python
e, v = np.linalg.eig(cx)
```

```python
plt.scatter(x[0,:], x[1,:], alpha=0.2)
for e_, v_ in zip(e, v.T):
    plt.plot([0, 3*e_*v_[0]], [0, 3*e_*v_[1]], 'r-', lw=2)
plt.axis([-3,3,-3,3]);
```

### Data matrices that have zero mean for all feature vectors

\begin{align}
\text{Cov}(X, Y) &= \frac{\sum_{i=1}^n(X_i - \bar{X})(Y_i - \bar{Y})}{n-1} \\
  &= \frac{\sum_{i=1}^nX_iY_i}{n-1} \\
  &= \frac{XY^T}{n-1}
\end{align}

and so the covariance matrix for a data set X that has zero mean in each feature vector is just $XX^T/(n-1)$. 

In other words, we can also get the eigendecomposition of the covariance matrix from the positive semi-definite matrix $XX^T$.

```python
e1, v1 = np.linalg.eig(np.dot(x, x.T)/(n-1))
```

```python
plt.scatter(x[0,:], x[1,:], alpha=0.2)
for e_, v_ in zip(e1, v1.T):
    plt.plot([0, 3*e_*v_[0]], [0, 3*e_*v_[1]], 'r-', lw=2)
plt.axis([-3,3,-3,3]);
```

### Using Singular Value Decomposition (SVD)

SVD is a decomposition of the data matrix $X = U S V^T$ where $U$ and $V$ are orthogonal matrices and $S$ is a diagnonal matrix. 

Recall that the transpose of an orthogonal matrix is also its inverse, so if we multiply on the right by $X^T$, we get the follwoing simplification

\begin{align}
X &= U S V^T \\
X X^T &= U S V^T (U S V^T)^T \\
 &= U S V^T V S U^T \\
 &= U S^2 U^T
\end{align}

Compare with the eigendecomposition of a matrix $A = W \Lambda W$, we see that SVD gives us the eigendecomposition of the matrix $XX^T$, which as we have just seen, is basically a scaled version of the covariance for a data matrix with zero mean, with the eigenvectors given by $U$ and eigenvealuse by $S^2$


```python
u, s, v = np.linalg.svd(x)
```

```python
e2 = s**2/(n-1)
v2 = u
plt.scatter(x[0,:], x[1,:], alpha=0.2)
for e_, v_ in zip(e2, v2):
    plt.plot([0, 3*e_*v_[0]], [0, 3*e_*v_[1]], 'r-', lw=2)
plt.axis([-3,3,-3,3]);
```

### We can transform the original data set so that the eigenvactors are the usual basis vectors

```python
ys = np.dot(v1.T, x)
```

```python
plt.scatter(ys[0,:], ys[1,:], alpha=0.2)
for e_, v_ in zip(e1, np.eye(2)):
    plt.plot([0, 3*e_*v_[0]], [0, 3*e_*v_[1]], 'r-', lw=2)
plt.axis([-3,3,-3,3]);
```

### And transform back to original coordinates

```python
zs = np.dot(v1, ys)
```

```python
plt.scatter(zs[0,:], zs[1,:], alpha=0.2)
for e_, v_ in zip(e1, v1.T):
    plt.plot([0, 3*e_*v_[0]], [0, 3*e_*v_[1]], 'r-', lw=2)
plt.axis([-3,3,-3,3]);
```
