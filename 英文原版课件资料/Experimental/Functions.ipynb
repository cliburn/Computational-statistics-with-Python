Estimating functions
----

- Interpolation: find a function $f$ that connects all points $(x_i, z_i)$ - i.e. $z_i = f(x_i)$
- Smoothing: find a function $f$ that minimizes the pointwise least squares sum $\sum_{i=1}^n (y_i - f(x_i))^2$
- Approximation: find a function $g$ that minimzes the distance $\left[ \int |f(x) - g(x)|^p w(x) dx \right]^{1/p}$

### Outline

- Function interpolation
    - polynomial interpolation
    - rational interpolation
    - Runges interpolation
    - Chebyshev intervals
    - Splines and the natural cubic spline
- Function smoothing
    - Splines again
    - Non-parametric methods
- Function approximation
    - Exact methods using power series, continued fractions and special functions
    - Inexact methods using polynomial, rational or other approximations

### Function interpolation

Since interpolation goes through all data points, it is most useful for noise-free or mostly noise-free systems.

#### Test function

```python
def runge(x):
    """Runges function."""
    return 1.0/(1+25*x**2)
```

```python
n = 100
x = np.linspace(-1, 1, n)
plt.plot(x, runge(x));
```

#### Polynomial interpolation

```python
# Simple polynomial interpolation is ill-conditioned

from scipy import interpolate

ix = np.linspace(-1, 1, 9)
f = interpolate.BarycentricInterpolator(ix, runge(ix))
plt.scatter(ix, f(ix))
plt.plot(x, runge(x), color='blue')
plt.plot(x, f(x), color='red');
```

```python
# Conditioning can be improved with use of apprpirately spaced sampling points

def chebyshev_nodes(n, a=-1, b=1):
    """Return Chebyshev nodes for polynomial interpolation."""
    return 0.5*(a+b) + 0.5*(b-a)*np.cos((2.0*np.arange(n)+1)/(2*n)*np.pi)
```

```python
ix = chebyshev_nodes(9)
f = interpolate.BarycentricInterpolator(ix, runge(ix))
plt.scatter(ix, f(ix))
plt.plot(x, runge(x), color='blue')
plt.plot(x, f(x), color='red');
```

#### Spline interpolation

```python
ix = np.linspace(-1, 1, 9)
f = interpolate.UnivariateSpline(ix, runge(ix), s=0) 
plt.scatter(ix, runge(ix))
plt.plot(x, runge(x), color='blue')
plt.plot(x, f(x), color='red');
```

#### Radial basis function interpolation

An example of 2D interpolation.

```python
from scipy.stats import multivariate_normal

def mixture(x):
    rv1 = multivariate_normal([-1,-1], 0.2*np.eye(2))
    rv2 = multivariate_normal([1,1], 0.3*np.eye(2))
    rv3 = multivariate_normal([-1,1], 0.5*np.eye(2))
    return 0.1*rv1.pdf(x) + 0.2*rv2.pdf(x) + 0.7*rv3.pdf(x)
```

```python
pts = np.random.uniform(-4,4,(100,2))
x, y = pts.T
z = mixture(pts)

f = interpolate.Rbf(x, y, z, epsilon=2)
```

```python
ti = np.linspace(pts.min(), pts.max(), 100)
XI, YI = np.meshgrid(ti, ti)
ZI = f(XI, YI)

plt.figure(figsize=(6,6))
plt.pcolor(XI, YI, ZI)
plt.scatter(x, y, 100, facecolor='none', alpha=0.5)
plt.axis([pts[:,0].min(), pts[:,0].max(), pts[:,1].min(), pts[:,1].max()])
plt.tight_layout()
```
