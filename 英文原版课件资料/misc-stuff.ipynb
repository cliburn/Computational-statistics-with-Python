```python
def quicksort(xs):
    if xs == []:
        return []
    else:
        smaller = [x for x in xs[1:] if x <= xs[0]]
        larger = [x for x in xs[1:] if x > xs[0]]
        return quicksort(smaller ) + [xs[0]] + quicksort(larger)
```

```python
xs = np.random.randint(0,10,10)
print xs
print quicksort(xs)
```
