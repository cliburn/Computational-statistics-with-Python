```python
def mandel_python(xs, ys, max_iters):
    c = np.zeros((len(xs), len(ys)), np.int)
    for i, x in enumerate(xs):
        for j, y in enumerate(ys):
            a = complex(x, y)
            z = 0.0j
            for k in range(max_iters):
                z = z*z + a
                if z.real*z.real + z.imag*z.imag >= 4:
                    c[i,j] = k
    return c
```

```python
xmin, xmax, ymin, ymax = -2, 1, -1, 1
xs = np.linspace(xmin, xmax, 200)
ys = np.linspace(ymin, ymax, 200)
```

```python
%timeit mandel_python(xs, ys, 200)
```

```python
c = mandel_python(xs, ys, 200)
plt.imshow(np.log1p(c.T), extent=[xmin, xmax, ymin, ymax]);
```

```python
%%cython -a

import numpy as np
cimport numpy as np
cimport cython

@cython.boundscheck(False)
@cython.wraparound(False)
def mandel_cython(double[::1] xs, double[::1] ys, int max_iters):

    m = xs.shape[0]
    n = ys.shape[0]

    cdef int[:,::1] c = np.zeros((m, n), np.int32)
    cdef int i, j, k
    cdef double complex a, z

    for i in range(m):
        for j in range(n):
            a.real = xs[i]
            a.imag = ys[j]
            z = 0.0j
            for k in range(max_iters):
                z = z*z + a
                if z.real*z.real + z.imag*z.imag >= 4:
                    c[i,j] = k
    return c
```

```python
%timeit mandel_cython(xs, ys, 200)
```

```python
c = mandel_cython(xs, ys, 200)
plt.imshow(np.log1p(c.T), extent=[xmin, xmax, ymin, ymax]);
```
